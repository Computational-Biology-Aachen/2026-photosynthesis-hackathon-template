# Cowpea IITA — Polars

This notebook analyses the cowpea dataset using **Polars** for fast offline-first analysis.

| | |
| --- | --- |
| **Crop** | Cowpea (_Vigna unguiculata_) |
| **Locations** | Ibadan, Ikenne, Kano (Nigeria) |
| **Years** | 2020–2022 |
| **Measurements** | 3,360 |
| **Genotypes** | 112 (98 with SNP markers) |
| **Tables** | `cowpea_iita_measurements`, `cowpea_iita_snp` |

```python
from pathlib import Path

import polars as pl

DATA = Path("..") / "data"

ms = pl.read_parquet(DATA / "cowpea_iita_measurements.parquet")
snp = pl.read_parquet(DATA / "cowpea_iita_snp.parquet")
```

## EDA

```python
print(f"Measurements: {ms.shape}")
print(f"SNP markers: {snp.shape}")
print(f"Genotypes: {ms['genotype'].n_unique()}")
print(f"Environments: {ms['env_id'].n_unique()}")
```

```python
# Counts by environment
ms.group_by("env_id", "location", "year", "stress_condition").len().sort("env_id")
```

```python
# Summary statistics for phenotype columns
phenotype_cols = [
    "relative_chlorophyll_content",
    "leaf_angle",
    "leaf_temp_differential",
    "lef",
    "npqt",
    "phi2",
    "phi_no",
    "phi_npq",
]

ms.select(phenotype_cols).describe()
```

```python
# Mean phenotype values by stress condition
ms.group_by("stress_condition").agg(
    pl.col("phi2").mean().alias("mean_phi2"),
    pl.col("lef").mean().alias("mean_lef"),
    pl.col("npqt").mean().alias("mean_npqt"),
    pl.col("relative_chlorophyll_content").mean().alias("mean_spad"),
    pl.col("leaf_temp_differential").mean().alias("mean_leaf_temp_diff"),
)
```

```python
# Mean phenotype values by location and stress
ms.group_by("location", "stress_condition").agg(
    pl.len().alias("n"),
    pl.col("phi2").mean().alias("mean_phi2"),
    pl.col("phi2").std().alias("std_phi2"),
    pl.col("lef").mean().alias("mean_lef"),
    pl.col("npqt").mean().alias("mean_npqt"),
).sort("location", "stress_condition")
```

```python
# Top 10 genotypes by mean Phi2
ms.group_by("genotype").agg(
    pl.len().alias("n"),
    pl.col("phi2").mean().alias("mean_phi2"),
    pl.col("lef").mean().alias("mean_lef"),
    pl.col("npqt").mean().alias("mean_npqt"),
).sort("mean_phi2", descending=True).head(10)
```

## Correlations

```python
# Pairwise Pearson correlations between phenotype traits
ms.select(phenotype_cols).drop_nulls().corr()
```

## SNP data exploration

```python
# Markers per chromosome
snp.group_by("chromosome").len().sort("chromosome")
```

```python
# Example: join mean phenotypes with SNP data
genotype_cols = [c for c in snp.columns if c.startswith("G")]

genotype_means = ms.group_by("genotype").agg(
    pl.col("phi2").mean().alias("mean_phi2"),
)

# Pick one marker and unpivot to long format
first_marker = snp.head(1)
marker_id = first_marker["marker_id"][0]

snp_long = first_marker.unpivot(
    on=genotype_cols,
    index="marker_id",
    variable_name="genotype",
    value_name="allele",
)

# Join with phenotype means
joined = snp_long.join(genotype_means, on="genotype")

print(f"Marker: {marker_id}")
joined.group_by("allele").agg(
    pl.len().alias("n"),
    pl.col("mean_phi2").mean().alias("phi2_by_allele"),
)
```

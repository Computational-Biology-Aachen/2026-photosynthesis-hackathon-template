## Stress response per genotype

Which genotypes maintain high Phi2 (photosynthetic efficiency) under drought stress?

```python
import matplotlib.pyplot as plt

# Mean phi2 per genotype under each stress condition
stress_by_geno = (
    ms.group_by("genotype", "stress_condition")
    .agg(pl.col("phi2").mean().alias("mean_phi2"))
    .pivot(on="stress_condition", index="genotype", values="mean_phi2")
    .rename({"Before Stress": "before", "During Stress": "during"})
    .with_columns((pl.col("before") - pl.col("during")).alias("stress_delta"))
    .sort("stress_delta")
)

print("Top 10 most stress-TOLERANT genotypes (smallest Phi2 drop):")
stress_by_geno.head(10)
```

```python
print("Top 10 most stress-SENSITIVE genotypes (largest Phi2 drop):")
stress_by_geno.tail(10).reverse()
```

```python
# Horizontal bar chart of stress delta for all genotypes
fig, ax = plt.subplots(figsize=(8, 18), layout="constrained")

df_plot = stress_by_geno.sort("stress_delta")
colors = [
    (
        "#2ca02c"
        if d <= df_plot["stress_delta"].quantile(0.25)
        else "#d62728" if d >= df_plot["stress_delta"].quantile(0.75) else "#7f7f7f"
    )
    for d in df_plot["stress_delta"].to_list()
]
ax.barh(df_plot["genotype"].to_list(), df_plot["stress_delta"].to_list(), color=colors)
ax.set_xlabel("Stress delta (Before - During mean Phi2)")
ax.set_ylabel("Genotype")
ax.set_title("Stress response per genotype\n(green = tolerant, red = sensitive)")
ax.axvline(0, color="black", linewidth=0.5)
ax.tick_params(axis="y", labelsize=6)
plt.show()
```

## Energy partitioning

Light energy absorbed by chlorophyll is partitioned into three pathways:
- **Phi2** — photochemistry (productive)
- **Phi_NPQ** — regulated non-photochemical quenching (photoprotection)
- **Phi_NO** — non-regulated dissipation (stress indicator)

These three fractions should sum to approximately 1.

```python
# Verify the energy balance: phi2 + phi_npq + phi_no ≈ 1
energy_check = ms.select(
    (pl.col("phi2") + pl.col("phi_npq") + pl.col("phi_no")).alias("energy_sum")
).drop_nulls()
print(
    f"Energy sum — mean: {energy_check['energy_sum'].mean():.4f}, "
    f"std: {energy_check['energy_sum'].std():.4f}, "
    f"min: {energy_check['energy_sum'].min():.4f}, "
    f"max: {energy_check['energy_sum'].max():.4f}"
)

# Mean fractions by stress condition
fractions_by_stress = (
    ms.group_by("stress_condition")
    .agg(
        pl.col("phi2").mean().alias("Phi2"),
        pl.col("phi_npq").mean().alias("Phi_NPQ"),
        pl.col("phi_no").mean().alias("Phi_NO"),
    )
    .sort("stress_condition")
)
print("\nMean energy fractions by stress condition:")
fractions_by_stress
```

```python
# Stacked bar chart: energy partitioning Before vs During stress
labels = fractions_by_stress["stress_condition"].to_list()
phi2_vals = fractions_by_stress["Phi2"].to_list()
phi_npq_vals = fractions_by_stress["Phi_NPQ"].to_list()
phi_no_vals = fractions_by_stress["Phi_NO"].to_list()

fig, ax = plt.subplots(figsize=(6, 5), layout="constrained")
x = range(len(labels))
ax.bar(x, phi2_vals, label="Phi2 (photochemistry)", color="#2ca02c")
ax.bar(
    x,
    phi_npq_vals,
    bottom=phi2_vals,
    label="Phi_NPQ (photoprotection)",
    color="#ff7f0e",
)
ax.bar(
    x,
    phi_no_vals,
    bottom=[a + b for a, b in zip(phi2_vals, phi_npq_vals)],
    label="Phi_NO (dissipation)",
    color="#d62728",
)
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylabel("Fraction of absorbed light")
ax.set_title("Energy partitioning by stress condition")
ax.legend(loc="upper right")
ax.set_ylim(0, 1.15)
ax.axhline(1.0, color="black", linewidth=0.5, linestyle="--", alpha=0.5)
plt.show()
```

```python
# Grouped stacked bars: location × stress condition
fractions_loc = (
    ms.group_by("location", "stress_condition")
    .agg(
        pl.col("phi2").mean().alias("Phi2"),
        pl.col("phi_npq").mean().alias("Phi_NPQ"),
        pl.col("phi_no").mean().alias("Phi_NO"),
    )
    .sort("location", "stress_condition")
)

locations = fractions_loc["location"].unique().sort().to_list()
conditions = ["Before Stress", "During Stress"]
bar_width = 0.35

fig, ax = plt.subplots(figsize=(9, 5), layout="constrained")

for i, cond in enumerate(conditions):
    subset = fractions_loc.filter(pl.col("stress_condition") == cond).sort("location")
    x_pos = [j + i * bar_width for j in range(len(locations))]
    p2 = subset["Phi2"].to_list()
    pn = subset["Phi_NPQ"].to_list()
    po = subset["Phi_NO"].to_list()

    ax.bar(
        x_pos,
        p2,
        bar_width,
        label=f"Phi2 ({cond})",
        color="#2ca02c",
        alpha=1.0 - 0.3 * i,
    )
    ax.bar(
        x_pos,
        pn,
        bar_width,
        bottom=p2,
        label=f"Phi_NPQ ({cond})",
        color="#ff7f0e",
        alpha=1.0 - 0.3 * i,
    )
    ax.bar(
        x_pos,
        po,
        bar_width,
        bottom=[a + b for a, b in zip(p2, pn)],
        label=f"Phi_NO ({cond})",
        color="#d62728",
        alpha=1.0 - 0.3 * i,
    )

ax.set_xticks([j + bar_width / 2 for j in range(len(locations))])
ax.set_xticklabels(locations)
ax.set_ylabel("Fraction of absorbed light")
ax.set_title("Energy partitioning by location and stress condition")
ax.legend(loc="upper right", fontsize=7, ncol=2)
ax.set_ylim(0, 1.15)
ax.axhline(1.0, color="black", linewidth=0.5, linestyle="--", alpha=0.5)
plt.show()
```

## Genome-wide association scan

For each of the 9,210 SNP markers, we ask: do genotypes with different alleles at this
locus differ in mean Phi2? We use the **range of allele-group means** as a simple effect
size metric (no external stats library needed).

```python
%%time

# Pre-compute genotype mean phi2
genotype_cols = [c for c in snp.columns if c.startswith("G")]
genotype_means = ms.group_by("genotype").agg(
    pl.col("phi2").mean().alias("mean_phi2"),
)

# Unpivot ALL markers to long format at once, then join with phenotype
snp_long_all = snp.unpivot(
    on=genotype_cols,
    index=["marker_id", "chromosome", "position"],
    variable_name="genotype",
    value_name="allele",
)

# Join with genotype means
snp_pheno = snp_long_all.join(genotype_means, on="genotype")

# Filter out missing allele calls (NN)
snp_pheno = snp_pheno.filter(pl.col("allele") != "NN")

# For each marker: compute range of mean_phi2 across allele groups
gwas = (
    snp_pheno
    .group_by("marker_id", "chromosome", "position", "allele")
    .agg(
        pl.col("mean_phi2").mean().alias("allele_mean_phi2"),
        pl.len().alias("n"),
    )
    .filter(pl.col("n") >= 5)  # require at least 5 genotypes per allele group
    .group_by("marker_id", "chromosome", "position")
    .agg(
        pl.col("allele_mean_phi2").max().alias("max_phi2"),
        pl.col("allele_mean_phi2").min().alias("min_phi2"),
        pl.col("allele").n_unique().alias("n_alleles"),
    )
    .filter(pl.col("n_alleles") >= 2)  # need at least 2 allele groups
    .with_columns(
        (pl.col("max_phi2") - pl.col("min_phi2")).alias("effect_size")
    )
    .sort("effect_size", descending=True)
)

print(f"Tested {gwas.height} markers (with ≥2 allele groups, ≥5 genotypes each)")
print("\nTop 20 markers by effect size:")
gwas.head(20)
```

```python
# Manhattan plot: effect size by genomic position, coloured by chromosome
chroms = sorted(gwas["chromosome"].unique().to_list())
cmap = plt.cm.get_cmap("tab20", len(chroms))
chrom_colors = {ch: cmap(i) for i, ch in enumerate(chroms)}

# Assign a cumulative x-position so chromosomes are laid out sequentially
chrom_offsets = {}
cum = 0
for ch in chroms:
    chrom_offsets[ch] = cum
    max_pos = gwas.filter(pl.col("chromosome") == ch)["position"].max()
    cum += max_pos + 1_000_000  # gap between chromosomes

gwas_plot = gwas.with_columns(
    pl.struct(["chromosome", "position"])
    .map_elements(
        lambda s: chrom_offsets[s["chromosome"]] + s["position"],
        return_dtype=pl.Int64,
    )
    .alias("x_pos")
)

fig, ax = plt.subplots(figsize=(14, 5), layout="constrained")
for ch in chroms:
    subset = gwas_plot.filter(pl.col("chromosome") == ch)
    ax.scatter(
        subset["x_pos"].to_list(),
        subset["effect_size"].to_list(),
        s=6,
        alpha=0.5,
        color=chrom_colors[ch],
        label=ch,
    )

# Chromosome labels at midpoints
tick_positions = []
tick_labels = []
for ch in chroms:
    ch_data = gwas_plot.filter(pl.col("chromosome") == ch)["x_pos"]
    mid = (ch_data.min() + ch_data.max()) / 2
    tick_positions.append(mid)
    tick_labels.append(ch.replace("VU", ""))
ax.set_xticks(tick_positions)
ax.set_xticklabels(tick_labels)

# Highlight threshold (top 1%)
threshold = gwas["effect_size"].quantile(0.99)
ax.axhline(
    threshold,
    color="red",
    linewidth=0.8,
    linestyle="--",
    label=f"Top 1% ({threshold:.4f})",
)

ax.set_xlabel("Chromosome")
ax.set_ylabel("Effect size (range of allele-group mean Phi2)")
ax.set_title("Manhattan plot — genome-wide association scan for Phi2")
ax.legend(fontsize=7, ncol=6, loc="upper right")
plt.show()
```

## Genotype × Environment interaction

A heatmap of mean Phi2 for each genotype in each environment reveals G×E interaction
patterns — genotypes that perform consistently well vs. those whose ranking changes
across environments.

```python
import numpy as np

# Build environment labels for nicer display
env_labels = (
    ms.group_by("env_id", "location", "year", "stress_condition")
    .len()
    .sort("env_id")
    .with_columns(
        (
            pl.col("location").str.slice(0, 3)
            + " "
            + pl.col("year").cast(pl.Utf8).str.slice(2)
            + " "
            + pl.when(pl.col("stress_condition") == "Before Stress")
            .then(pl.lit("BS"))
            .otherwise(pl.lit("DS"))
        ).alias("env_label")
    )
)

# Pivot: genotype × env_id, values = mean phi2
gxe = (
    ms.group_by("genotype", "env_id")
    .agg(pl.col("phi2").mean().alias("mean_phi2"))
    .pivot(on="env_id", index="genotype", values="mean_phi2")
    .sort("genotype")
)

# Reorder columns by env_id
env_ids_sorted = sorted([c for c in gxe.columns if c != "genotype"], key=int)
gxe = gxe.select(["genotype"] + env_ids_sorted)

# Sort genotypes by overall mean phi2 for a meaningful ordering
geno_order = (
    ms.group_by("genotype")
    .agg(pl.col("phi2").mean().alias("overall_mean"))
    .sort("overall_mean", descending=True)["genotype"]
    .to_list()
)
gxe = gxe.with_columns(pl.col("genotype").cast(pl.Categorical).cast(pl.Utf8))
gxe_sorted = pl.concat([gxe.filter(pl.col("genotype") == g) for g in geno_order])

# Convert to numpy for imshow
geno_names = gxe_sorted["genotype"].to_list()
matrix = gxe_sorted.select(env_ids_sorted).to_numpy()

# Environment labels for x-axis
env_label_map = dict(
    zip(
        env_labels["env_id"].cast(pl.Utf8).to_list(),
        env_labels["env_label"].to_list(),
    )
)
x_labels = [env_label_map.get(e, e) for e in env_ids_sorted]

fig, ax = plt.subplots(figsize=(10, 18), layout="constrained")
im = ax.imshow(matrix, aspect="auto", cmap="RdYlGn", interpolation="nearest")
ax.set_xticks(range(len(x_labels)))
ax.set_xticklabels(x_labels, rotation=45, ha="right", fontsize=8)
ax.set_yticks(range(len(geno_names)))
ax.set_yticklabels(geno_names, fontsize=5)
ax.set_xlabel("Environment")
ax.set_ylabel("Genotype (sorted by overall mean Phi2)")
ax.set_title("Genotype × Environment interaction — mean Phi2")
cbar = fig.colorbar(im, ax=ax, shrink=0.5)
cbar.set_label("Mean Phi2")
plt.show()
```
